### Dataset Overview
For this analysis, we are utilizing a Melbourne housing dataset. This dataset provides a snapshot of real estate transaction data from 2017 in the Melbourne metropolitan area. A cleaned version of the dataset was available on [Kaggle](https://www.kaggle.com/datasets/dansbecker/melbourne-housing-snapshot/data), but the original dataset was scraped from publicly available listings posted weekly on [Domain.com.au](Domain.com.au). It contains information on residential properties including their characteristics, like number of bedrooms and plot size, as well as location, and sale price.

### Research Question
The primary objective of this analysis is to develop a predictive model capable of estimating housing prices based on property characteristics. Specifically, we aim to answer: **What property features most strongly influence housing prices in Melbourne, and how well can these features accurately predict the market value of a given property?**

### Key Features Selected for Analysis
Based on the available data and the relevance to price prediction, we have selected the following features:

- **Rooms**: Number of bedrooms
- **Bathroom**: Number of bathrooms
- **Car**: Number of parking spaces
- **Landsize**: Plot size of the property
- **BuildingArea**: Total building area
- **YearBuilt**: Year the property was constructed
- **Propertycount**: Count of properties in the suburb

### Feature Selection Rationale
These features were selected because they directly represent physical and locational characteristics that historically influence property valuation. Features such as Address, Method of sale, and Agent information were excluded as they do not represent intrinsic property characteristics. Geographical coordinates (Latitude/Longitude) and administrative codes were similarly excluded to avoid multicollinearity while preserving predictive power through the included feature set.

### Target Variable
The target variable for our prediction model is **Price**, representing the sale price of each property in the dataset.
